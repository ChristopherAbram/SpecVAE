# Analyze SpecVAE score

Different alpha parameters are investigated

```python
import os
import numpy as np
from matplotlib import pyplot as plt

import specvae.utils as utils
from specvae.utils import filter_score_matrix
import specvae.dataset as dt
from specvae.visualize import plot_distribution, plot_precentile
from specvae.dataset import MoNA
```

```python
dataset = 'MoNA'
```

## SpecVAE Euclidean

```python
jaccard_scores = np.load(utils.get_project_path() / '.data' / dataset / ('%s_sim_jaccard.npy' % dataset), allow_pickle=True)
vs = np.load(utils.get_project_path() / '.data' / dataset / ('%s_vaesim_eu.npz' % dataset), allow_pickle=True)
vae_eu_sim = vs['vae_eu']
```

```python
ignore_diagonal = False
num_bins = 100
```

```python
eu_per = []
for sim in vae_eu_sim:
    eu_per.append(plot_precentile(jaccard_scores, sim, num_bins=num_bins, 
        show_top_percentile=1.0, ignore_diagonal=False))
```

```python
for sim in vae_eu_sim:
    plot_precentile(jaccard_scores, sim, num_bins=num_bins, 
        show_top_percentile=0.1, ignore_diagonal=False)
```

```python
for sim in vae_eu_sim:
    plot_precentile(jaccard_scores, sim, num_bins=num_bins, 
        show_top_percentile=1.0, ignore_diagonal=True)
```

## SpecVAE KL-divergence

```python
vs = np.load(utils.get_project_path() / '.data' / dataset / ('%s_vaesim_kl.npz' % dataset), allow_pickle=True)
vae_kl_sim = vs['vae_kl']
```

```python
kl_per = []
for sim in vae_kl_sim:
    kl_per.append(plot_precentile(jaccard_scores, sim, num_bins=num_bins, 
        show_top_percentile=1.0, ignore_diagonal=False))
```

```python
for sim in vae_kl_sim:
    plot_precentile(jaccard_scores, sim, num_bins=num_bins, 
        show_top_percentile=0.1, ignore_diagonal=False)
```

```python
for sim in vae_kl_sim:
    plot_precentile(jaccard_scores, sim, num_bins=num_bins, 
        show_top_percentile=1.0, ignore_diagonal=True)
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(7,6))
x_percentiles = (show_top_percentile / num_bins * (1 + np.arange(num_bins)))[::-1]

for per in eu_per:
    plt.plot(x_percentiles, per)
plt.show()
```
